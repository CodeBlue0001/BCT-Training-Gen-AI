2. **Context-Aware Documentation Search**
   * **Concept:** Use Python to scrape or read local `.md` or `.py` files, and use the LLM to answer questions specifically based on that code documentation.
   * **Why it's valued:** Demonstrates knowledge of RAG (Retrieval-Augmented Generation) concepts and data handling.

3. **Specialized Technical Interview Prep Bot**
   * **Concept:** Modify your current chatbot to act as a 'Python Interviewer'. It asks the user a technical question, evaluates the user's Python code answer, and provides constructive feedback.
   * **Why it's valued:** Shows you can design complex system prompts and handle logical workflows in Python.

```python
# Example: Quick implementation of the 'Code Reviewer' idea
def ai_code_reviewer(code_snippet):
    reviewer_chat = client.chats.create(
        model="gemini-2.5-flash",
        config=types.GenerateContentConfig(
            system_instruction="You are a Senior Developer. Review the provided code for bugs, efficiency, and PEP 8 style. Provide a brief 15 to 20 line summary with time and space complexity.If there is any bug that will be marked and will be corrected."
        )
    )
    response = reviewer_chat.send_message(f"Review this code:\n\n{code_snippet}")
    return response.text

# Test
sample_code = "def add(a,b): return a+b"
# print(ai_code_reviewer(sample_code))
```

###  Project 1.1: Live AI Screening Simulation

To make this impactful for a resume, you can frame this as a **'Real-time Developer Feedback Loop'**.

**Resume Bullet Point:**
> *Developed a real-time code analysis engine using Gemini Pro and Python, reducing debugging time by providing instantaneous PEP 8 and logic feedback during active coding sessions.*

```python
from flask import Flask, render_template_string, request, jsonify
from google.colab import output
import threading

app = Flask(__name__)

# HTML/CSS/JS for a professional Code Editor interface
HTML_TEMPLATE = """
<!DOCTYPE html>
<html>
<head>
    <title>AI Code Reviewer</title>
    <style>
        body { font-family: 'Segoe UI', sans-serif; background: #1e1e1e; color: #fff; margin: 0; padding: 20px; }
        .container { display: flex; gap: 20px; height: 60vh; }
        .editor-section, .feedback-section { flex: 1; display: flex; flex-direction: column; }
        textarea { flex: 1; background: #2d2d2d; color: #d4d4d4; font-family: monospace; padding: 15px; border: 1px solid #444; border_radius: 5px; resize: none; font-size: 14px; }
        button { background: #007acc; color: white; border: none; padding: 12px; margin-top: 10px; cursor: pointer; border-radius: 5px; font-weight: bold; }
        button:hover { background: #005f9e; }
        .feedback-box { flex: 1; background: #252526; padding: 15px; border: 1px solid #444; border-radius: 5px; overflow-y: auto; white-space: pre-wrap; font-size: 14px; color: #cecece; }
        h2 { margin-top: 0; color: #569cd6; }
        #loading { display: none; color: #dcdcaa; margin-bottom: 10px; }
    </style>
</head>
<body>
    <h2>🚀 AI Live Code Screening</h2>
    <div class="container">
        <div class="editor-section">
            <textarea id="codeEditor" placeholder="Write your Python code here..."></textarea>
            <button onclick="analyzeCode()">Analyze Code</button>
        </div>
        <div class="feedback-section">
            <div id="loading">⏳ Analyzing...</div>
            <div id="feedback" class="feedback-box">Waiting for input...</div>
        </div>
    </div>

    <script>
        async function analyzeCode() {
            const code = document.getElementById('codeEditor').value;
            const feedback = document.getElementById('feedback');
            const loading = document.getElementById('loading');

            if(!code.trim()) return;

            loading.style.display = 'block';
            feedback.innerText = '';

            const response = await fetch('/review', {
                method: 'POST',
                headers: {'Content-Type': 'application/json'},
                body: JSON.stringify({code: code})
            });
            const data = await response.json();

            loading.style.display = 'none';
            feedback.innerText = data.review;
        }
    </script>
</body>
</html>
"""

@app.route('/')
def index():
    return render_template_string(HTML_TEMPLATE)

@app.route('/review', methods=['POST'])
def review():
    code = request.json.get('code')
    try:
        # Setup specialized reviewer

        response = ai_code_reviewer.send_message(f"Review this code:\n{code}")
        return jsonify({'review': response.text})
    except Exception as e:
        return jsonify({'review': f"Error: {str(e)}"})

def run_app():
    app.run(port=5000)

# Run Flask in a background thread
threading.Thread(target=run_app).start()

# Show the app in the output cell
output.serve_kernel_port_as_iframe(5000, height='600')
```
